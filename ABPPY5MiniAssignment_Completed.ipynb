##1

```python
import pandas as pd

# First, load the IceCreamData.csv file into a DataFrame and check the result with the .head() and .tail()
df = pd.read_csv('IceCreamData.csv')

print("Head:")
print(df.head())

print("\nTail:")
print(df.tail())
```

```python
# Second, join the New ice cream data so that it is at the bottom of the original Ice Cream Data.
# Read in the new data and add those rows to your original DataFrame.
# How many rows are in the new table? Check the result with the .tail() and also find the len() of the new DataFrame

df_new = pd.read_csv('IceCreamDataNew.csv')

df = pd.concat([df, df_new], ignore_index=True)

print("Tail after concatenation:")
print(df.tail())

print("\nNumber of rows:", len(df))
```

```python
# Finally, merge the data using the Customer ID with the last file: CustomerData.csv

df_customers = pd.read_csv('CustomerData.csv')

df = pd.merge(df, df_customers, on='Customer', how='left')

print(df.head())
print(df.shape)
```

##2

```python
# Remove the Customer column
df = df.drop(columns=['Customer'])
print(df.head())
```

```python
# Create a summary grouping by Flavor showing means of Age
print(df.groupby('Flavor')['Age'].mean())
```

```python
# Create a summary grouping by Store showing means of Sales
print(df.groupby('Store')['Sales'].mean())
```

```python
# Create a summary grouping by Gender showing means of Age
print(df.groupby('Gender')['Age'].mean())
```

```python
# Create a summary grouping by Gender showing medians of Age
print(df.groupby('Gender')['Age'].median())
```

```python
# Create a summary grouping by Flavor showing means of Age and Sales (the only quant variables)
print(df.groupby('Flavor')[['Age', 'Sales']].mean())
```

```python
# Create the above summary so that the means have the prefix 'mean_'
summary = df.groupby('Flavor')[['Age', 'Sales']].mean()
summary.columns = ['mean_' + col for col in summary.columns]
print(summary)
```

```python
# Create a new summary that groups by Flavor and then Type and shows the mean of Sales
print(df.groupby(['Flavor', 'Type'])['Sales'].mean())
```

```python
# Create a summary that groups by Flavor and returns the Age and Sales mean and count
print(df.groupby('Flavor')[['Age', 'Sales']].agg(['mean', 'count']))
```

```python
# Create a summary that groups by Flavor and returns the mean and standard deviation of Sales.
# But make sure to use the friendly titles "Average" and "Standard Deviation"
summary = df.groupby('Flavor')['Sales'].agg(
    Average='mean',
    **{'Standard Deviation': 'std'}
)
print(summary)
```

```python
# Finally create a pivot table that returns the counts with Store as the rows
# and Type as the columns. Include the margin totals
pivot = pd.pivot_table(df, values='Sales', index='Store', columns='Type',
                       aggfunc='count', margins=True, margins_name='Total')
print(pivot)
```

```python
# Save all the qualitative columns in a list. Remove the Name columns.
# You can use slicing to do this. Hint: [0:?]
# Then loop through the list and output the .value_counts() for each column

# All columns: Type, Flavor, Store, Gender, FirstName, LastName, Age, Sales
# Qualitative (object dtype) columns — slice off the Name columns (FirstName, LastName)
qual_cols = list(df.select_dtypes(include='object').columns)  # ['Type','Flavor','Store','Gender','FirstName','LastName']
qual_cols = qual_cols[0:4]  # Remove FirstName and LastName, keep Type, Flavor, Store, Gender

print("Qualitative columns:", qual_cols)
print()

for col in qual_cols:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts())
    print()
```

```python
# Save all the quantitative columns in a list.
# Then loop through the list and output the .describe() for each column

quant_cols = list(df.select_dtypes(include='number').columns)  # ['Age', 'Sales']

print("Quantitative columns:", quant_cols)
print()

for col in quant_cols:
    print(f"Descriptive statistics for '{col}':")
    print(df[col].describe())
    print()
```

```python
# This function should have two inputs:
#   a dataframe
#   a code which should be either "N" for number or "C" for categorical
#   Based on the entered value, return the appropriate summaries (similar to above)

def column_summaries(dataframe, code):
    if code == 'N':
        quant_cols = list(dataframe.select_dtypes(include='number').columns)
        print("Quantitative columns:", quant_cols)
        print()
        for col in quant_cols:
            print(f"Descriptive statistics for '{col}':")
            print(dataframe[col].describe())
            print()
    elif code == 'C':
        qual_cols = list(dataframe.select_dtypes(include='object').columns)
        qual_cols = qual_cols[0:4]  # Remove Name columns (FirstName, LastName)
        print("Categorical columns:", qual_cols)
        print()
        for col in qual_cols:
            print(f"Value counts for '{col}':")
            print(dataframe[col].value_counts())
            print()
    else:
        print("Invalid code. Please enter 'N' for number or 'C' for categorical.")


# Test the function
print("=== Testing with code 'N' ===")
column_summaries(df, 'N')

print("=== Testing with code 'C' ===")
column_summaries(df, 'C')
```
