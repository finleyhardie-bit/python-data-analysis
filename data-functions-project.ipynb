```python
def create_mean(series):
    mean_val = series.mean()
    col_name = series.name if series.name is not None else "entered data"
    return f"The mean of the {col_name} is: {mean_val:.2f}"

create_mean(df_IC["Age"])
```

```python
def create_value_counts(series):
    vc = series.value_counts()
    col_name = series.name if series.name is not None else "entered data"
    
    parts = ""
    for label, count in vc.items():
        parts += f"{label}: {count} "
    
    parts = parts.strip()
    
    return f"The {col_name} counts are: {parts}"

create_value_counts(df_IC["Type"])
```
